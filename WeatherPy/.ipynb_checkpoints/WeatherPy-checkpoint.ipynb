# WeatherPy

In this example, you'll be creating a Python script to visualize the weather of 500+ cities across the world of varying distance from the equator. To accomplish this, you'll be utilizing a [simple Python library](https://pypi.python.org/pypi/citipy), the [OpenWeatherMap API](https://openweathermap.org/api), and a little common sense to create a representative model of weather across world cities.

Your objective is to build a series of scatter plots to showcase the following relationships:

* Temperature (F) vs. Latitude
* Humidity (%) vs. Latitude
* Cloudiness (%) vs. Latitude
* Wind Speed (mph) vs. Latitude


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from pprint import pprint
from datetime import datetime

# import the API key for the Openweathermap web page
from api_keys import api_key

# inport citipy to determine city based on latitude and longitude
from citipy import citipy

# output file (CSV)
output_data_file = "output_data/cities.csv"

# range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# define lists to hold reponse info and with global scope
cloudiness, country, date, humidity = [], [], [], []
lat, lon, max_temp, wind_speed      = [], [], [], [] 
cities_retrieved                    = []          # identifies cities for which valid data is retrieved 

# define a time variable with global scope, which is used in the plots
current_date = '00/00/00'
```

## Generate Cities List

```python
# define list for holding lat_lngs and cities
lat_lngs = []
cities = []

# create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# identify the nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # if the city is unique, then add it to the cities list
    if city not in cities:
        cities.append(city)

# confirm the number of cities is sufficient
print(f"Number of cities = {len(cities)}")
```

#### Perform a weather check on each city using a series of successive API calls.

```python
# save config information
url = "http://api.openweathermap.org/data/2.5/weather?"

# build the query URL
query_url = url + "appid=" + api_key + "&q="

# specify set sizes
if len(cities) > 50:
    set_size = 50
    number_sets = len(cities) // set_size
else:
    set_size = len(cities)
    number_sets = 1    

# reset the current_date to indicate that a current_date has not been calculated, yet
current_date = '00/00/00'

# get weather data for all of the listed cities
print("Beginning Data Retrieval")     
print("-----------------------------")

for i in range(number_sets):
    for j in range(set_size):
        try:
            weather_json = requests.get(query_url + cities[i+j]).json()
            cloudiness.append(weather_json['clouds']['all'])
            country.append(weather_json['sys']['country'])
            date.append(weather_json['dt'])
            humidity.append(weather_json['main']['humidity'])
            lat.append(weather_json['coord']['lat'])
            lon.append(weather_json['coord']['lon'])
            max_temp.append(weather_json['main']['temp_max'])
            wind_speed.append(weather_json['wind']['speed'])
            cities_retrieved.append(cities[i+j])
            print(f"Processing Record {j+1} of Set {i+1} | {cities[i+j]}")            
            
            if current_date == '00/00/00':   # only needed once in this loop
                # convert the retrieved UNIX UTC time to mm/dd/yy format, which is used in plot titles
                ts = int(weather_json['dt'])
                current_date = datetime.utcfromtimestamp(ts).strftime('%m/%d/%y')
        except:
            print("#####  City not found. Skipping...")
            pass

print("-----------------------------")
print("End of Data Retrieval")     
print()
print(f"The current date is {current_date}")
print(f"Weather for {len(max_temp)} out of {len(cities)} cities is retrieved.")

# convert Kelvin temperature to Fahrenheit, rounded to two decimal places, using list comprehension
max_temp = [round((float(t) * 9/5) - 459.67, 2) for t in max_temp]
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# insure the data lists are all of the same length, otherwise the data frame will not form
print(f"  Count of cities retrieved:  {len(cities_retrieved)}")
print(f"Count of cloudiness reports:  {len(cloudiness)}"      )
print(f"   Count of country reports:  {len(country)}"         )
print(f"      Count of date reports:  {len(date)}"            )
print(f"  Count of humidity reports:  {len(humidity)}"        )
print(f"       Count of lat reports:  {len(lat)}"             )
print(f"       Count of lon reports:  {len(lon)}"             )
print(f"  Count of max_temp reports:  {len(max_temp)}"        )
print(f"Count of wind_speed reports:  {len(wind_speed)}"      )
```

```python
# create a data frame for the weather data received
weather_dict = {
    "City": cities_retrieved,
    "Cloudiness": cloudiness, "Country": country, "Date": date, "Humidity": humidity, 
    "Lat": lat, "Lng": lon, "Max Temp": max_temp, "Wind Speed": wind_speed
}
weather_data = pd.DataFrame(weather_dict)
weather_data.to_csv(output_data_file, index=False)
weather_data
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

```python
# the required plots are the same, with only the y-variable changing... so a function is defined to plot each
def plot_data(x, y, y_label, title):
    plt.scatter(x, y, color='cornflowerblue', marker='o', alpha = .2, edgecolors='k')
    plt.xlabel('Latitude')
    plt.ylabel(y_label)
    plt.title(title + "  (" + current_date + ")")
    plt.grid(True)
    plt.savefig("output_data/" + title + ".png")
    plt.show()
```

#### Latitude vs. Temperature Plot

```python
plot_data(lat, max_temp, 'Max Temperature (F)', 'City Latitude vs. Max Temperature')
```

#### Latitude vs. Humidity Plot

```python
plot_data(lat, humidity, 'Humidity (%)', 'City Latitude vs. Humidity')
```

#### Latitude vs. Cloudiness Plot

```python
plot_data(lat, cloudiness, 'Cloudiness (%)', 'City Latitude vs. Cloudiness')
```

#### Latitude vs. Wind Speed Plot

```python
plot_data(lat, wind_speed, 'Wind Speed (mph)', 'City Latitude vs. Wind Speed')
```

## Trend Analysis
1. It is summer now in the southern hemisphere and winter in the northern. Summer temperatures are warmer than the winter temperatures.
2. Temperature strongly decreases with distance from the equator in the hemisphere that is in winter.
3. Temperature does not change with latitude near the equator in the hemisphere that is in summer, and then starts to sharply decrease when the latitude is beyond the 45th parallel.
4. Wind speed tends to increase  and then decrease with distance from the equator (It's windiest around the 45th parallel.)
5. Humidity and cloudiness are not affected by latitude. (It rains on both sinner and saint, alike.)
